```python
filtred_ratings = ratings.groupby(['userId']).agg({'rating':['count'], 'timestamp':['max', 'min']})
```

```python
filtred_ratings['lifetime'] = filtred_ratings['timestamp']['max'] - filtred_ratings['timestamp']['min']
```

```python
res_ratings = filtred_ratings.loc[filtred_ratings['rating']['count'] > 100]
```

```python
mean_lifetime = res_ratings['lifetime'].mean()

print(f'Среднее время жизни пользователей, которые выставили более 100 оценок = {int(mean_lifetime)} сек')
print(f'Среднее время жизни пользователей, которые выставили более 100 оценок = {int(mean_lifetime) / 3600 / 24:.0f} дня')
```

```python
# Вариант 3 - groupby() + apply()
```

```python
ratings = ratings.groupby('userId').agg({'rating': 'count', 'timestamp': ['min', 'max']}).reset_index()
```

```python
ratings = ratings.loc[ratings['rating']['count'] > 100].apply(lambda row: row['timestamp']['max'] - row['timestamp']['min'], axis=1)
ratings.head(10)
```

```python
mean_lifetime = ratings.iloc[:].mean().round().astype(int)


print(f'Среднее время жизни пользователей, которые выставили более 100 оценок = {int(mean_lifetime)} сек')
print(f'Среднее время жизни пользователей, которые выставили более 100 оценок = {int(mean_lifetime) / 3600 / 24:.0f} дня')
```

таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
аналогичную таблицу по типам выручки с указанием адреса клиента

```python
# Обратите внимание, что в процессе объединения таблиц данные не должны теряться.
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
joined_1 = rzd.merge(auto, on='client_id', how='outer').merge(air, on='client_id', how='outer').fillna(0)
joined_1.head(10)
```

```python
joined_2 = client_base.merge(joined_1, on='client_id', how='left')
joined_2.head(10)
```

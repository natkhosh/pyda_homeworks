#### 2. Постройте график функции, первой и второй производной:

#### На отрезке [-5; 5].

```python
x = sym.Symbol('x')
```

```python
# Первая производная функции

diff_2_1 = sym.diff(sym.sin(2 * x + 1)**5)
diff_2_1
```

```python
# Второя производная функции

diff_2_2 = sym.diff(diff_2_1)
diff_2_2
```

```python
# создаём массив данных для переменной х
X = np.linspace(-5, 5)
X
```

```python
# строим график первой производной функции f(x)
```

```python
def func_diff_2_1(x):
    return 10 * np.sin(2 * x + 1)** 4 * np.cos(2 * x + 1)
```

```python
plt.plot(X, func_diff_2_1(X))
plt.xlabel("x = [-5, 5]")
plt.ylabel("f(x)")
plt.title('Первая производная f(x)')
plt.grid()
plt.show()
```

```python
# строим график второй производной функции f(x)
```

```python
def func_diff_2_2(x):
    return -20 * np.sin(2 * x + 1)** 5 + 80 * np.sin(2 * x + 1)** 3 * np.cos(2 * x + 1)** 2
```

```python
plt.plot(X, func_diff_2_2(X))
plt.xlabel("x = [-5, 5]")
plt.ylabel("f(x)")
plt.title('Вторая производная f(x)')
plt.grid()
plt.show()
```

### Задание 1

#### Создайте случайную величину из выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистограммы средних этих выборок.

```python
expon_rv = stats.expon(scale =1.5)
sample = expon_rv.rvs(size=1000)
```

```python
x = np.linspace(0,5,100)
```

```python
pdf = expon_rv.pdf(x)

plt.figure(figsize=(10,8))
plt.hist(sample, density=True, bins=50)
plt.plot(x, pdf, alpha=0.8)

plt.show()
```

```python
# гистограмма с помощью библиотеки seaborn
sns.histplot(x=sample, kde=True)
```

### Задание 2

#### При каком значение n полученное распредление похоже на нормальное?

```python
samples_count = 1000

# Массивы для сохранения средних по выборкам
for smp in [2, 10, 100, 1000, 10000]:
    mList = []
    
    for kk in range(samples_count):
        # На каждом цикле генерируем выборки разных размеров
        data = expon_rv.rvs(size=smp)
        # Считаем среднее по каждой выборке и добавляем в массив средних
        mList.append(data.mean())
          
    sns.histplot(x=mList, kde=True)
    plt.show()
```

### Задание 1

#### Монетку подбрасывают 15 раз, найдите вероятность выпадения простого числа орлов.

##### Для решения задачи воспользуемся формулой Бернулли:


```python
# количество подбрасываний монетки
n = 15
```

```python
# Простые числа до 15
simple_values = [2, 3, 5, 7, 11, 13] 
```

```python
P = 0
for k in simple_values:
    C = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    P += C * (0.5 ** n)

print(f"Вероятность выпадения простого числа орлов: {P:.2f} или {P:2.0%}")
```

### Задание 2

#### Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
from scipy.stats import norm
```

```python
import scipy.stats as stat
```

```python
xmin = -10
xmax = 30
ymax = 0.6

x = np.linspace(xmin, xmax, num=200000)  
```

```python
plt.figure(figsize=(12,12))

# Параметры :
    # loc — параметр сдвига;
    # scale — параметр масштаба;

plt.plot(x, stat.norm.pdf(x, loc=0, scale=1))
plt.plot(x, stat.norm.pdf(x, loc=4, scale=4))
plt.plot(x, stat.norm.pdf(x, loc=8, scale=0.8))
plt.plot(x, stat.norm.pdf(x, loc=10, scale=1.5))
plt.plot(x, stat.norm.pdf(x, loc=13, scale=5))

plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.xlim((xmin, xmax))
plt.ylim((None, ymax))
plt.show()  

```
